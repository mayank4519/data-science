```python
# we need a new data frame having 3 columns
# region, Happiness Rank Per Region, numCountriesInRegion
```

```python
df.head()
```

```python
regions = df['Region'].unique()
happiness_rank = []
countries_per_region = []
row_list = []


for region in regions:
    happiness_rank = ((df[df['Region'] == region]['Happiness Rank'].sum())/len(df[df['Region'] == region]))
    countries_per_region = (len(df[df['Region'] == region]))
    
    dict1 = {'Region': region,
                'Happiness Rank Per Region': happiness_rank,
                'numCountriesInRegion': countries_per_region}
    
    row_list.append(dict1)
    
    
new_df = pd.DataFrame(row_list)
new_df.sort_values(by = 'Happiness Rank Per Region', ascending=True, inplace=True)
```

```python
new_df
```

```python
plt.figure(figsize=(12,10))
ax = sns.barplot(new_df['Region'],new_df['Happiness Rank Per Region'])
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, ha = 'right')
```
